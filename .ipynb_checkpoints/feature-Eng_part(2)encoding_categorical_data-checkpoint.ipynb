##### Encoding Categorical Data
        -Nominal  => Categories with no natural order i.e Country → USA, UK, Pak
        -Ordinal => Categories with a meaningful order, but the difference between categories is not necessarily equal i.e Education Level → High School <                     Bachelor < Master < PhD

### two Techniques use => Ordinal encoding and one hot encoding

#### Label encoding and ordinal encoding

 ### Label encoding
        when we have y column mean output column is categorical so we use label encoding
        when we have x  mean input columns are categorical so we use ordinal encoding
