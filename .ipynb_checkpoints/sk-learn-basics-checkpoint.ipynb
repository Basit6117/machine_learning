```python
#preprocessing
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
cust = pd.read_csv('customer.csv')
cust.head(8)
```

```python
y = cust['purchased']
label_enc = LabelEncoder()
y_enc = label_enc.fit_transform(y)
```

```python
y_enc
```

```python
cust['purchased'] = y_enc
```

```python
cust.sample(10)
```

```python
ohe = OneHotEncoder()
encoded_data = ohe.fit_transform(cust[['review']])
encoded_df = pd.DataFrame(
    encoded_data.toarray(),
    columns=ohe.get_feature_names_out(['review'])
)

```

```python
encoded_df
```

### Estimator

    In machine learning, an estimator is a function or algorithm that learns from observed data to approximate unknown parameters or functions of a population.
    estimators will always have fit() method

#### Two types of estimator
    - predictor (an estimator that predict after learning from data through predic() mathed )
    - transformer (transformation after learning from data through transform()/ fit_transform())
