### Categorical data

#### a. Countplot

```python
df = pd.read_csv('train.csv')
df
```

```python
sns.countplot(x='Survived', data=df) # countplot() => frequency count of each category
```

```python
sns.countplot(x='Sex', data=df)
```

```python
### Pie chart
```

```python
# df['Sex'].value_counts().plot(kind='pie', autopct='%.2f')
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f')
```
