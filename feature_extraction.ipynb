# Feature Extraction

#### Curse of Dimentionality
    1. When number of features (dimensions) increases, data becomes sparse (spread out).
    2. Machine learning models need more data to understand high-dimensional space.
    3. Distance-based algorithms (KNN, K-Means) become unreliable in high dimensions.
    4. The model becomes slower and overfits easily.
    5. Noise increases, useful information decreases.
    6. Visualization becomes difficult because you can’t “see” data beyond 3D.
    7. Models perform worse unless we do feature selection or dimensionality reduction (PCA).


#### What is “dimension”?
    Dimension = number of features (columns) in your dataset.
#### Example:
        Age (1D)
        Age + Salary (2D)
        Age + Salary + Height (3D)
        100 features → 100D space    

## Dimentionality reduction techniques
    - Feature selection   
    - Feature extraction
    

# Feature Extraction Techniques

### PCA
    it is a an unsupervised machine learning problem (data have only inputs no output)
    PCA is used to reduce curse of dimentionality
    PCA bring high dementional data into lower dimentional data

#### Benefits of PCA
    - Faster execution of algorithm (because of dimentionality reduction)
    - Visualizaton (i.e bring the 10D into 3D or 2D )
