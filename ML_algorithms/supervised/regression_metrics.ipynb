# **MAE â€” Mean Absolute Error**

### **ðŸ”¹ Simple Definition**

**MAE measures the average absolute difference between the predicted value and the actual value.**

It tells you:

> â€œOn average, how far are your predictions from the true values?â€

---

# **Formula**

$MAE = \frac{1}{m} \sum_{i=1}^{m} |y^{(i)} - \hat{y}^{(i)}|$

**Where:**  

- $$y^{(i)}$$ = actual value    
- $$\hat{y}^{(i)}$$ = predicted value    
- $$| \cdot |$$ = absolute difference    
- $$m$$ = number of samples    

---

# ðŸ”¹ **Key Idea**

* Take the error (difference).
* Ignore +/- sign (absolute).
* Average them.

---

# **MAE Example (very simple)**

Actual values: [10, 20, 30]
Predicted values: [12, 18, 25]

Errors:

* |10 â€“ 12| = 2
* |20 â€“ 18| = 2
* |30 â€“ 25| = 5

MAE = (2 + 2 + 5) / 3 = **3**

This means:
ðŸ‘‰ *On average, predictions are 3 units away from the true values.*

---

# **Advantages of MAE**

1. **Easy to understand**
2. **Less sensitive to outliers** than MSE
3. Represents real-world average error
4. Useful when all errors should be treated equally

---

# **Disadvantages of MAE**

1. Doesn't heavily penalize big errors
2. Not smooth for optimization (absolute values are harder to derive)

---

# When to use MAE?

* When you want a **simple, interpretable metric**
* When your dataset has **outliers**
* When you want to measure average prediction error directly

# **Mean Squared Error (MSE)**

### **1. Definition**

* MSE measures the **average of the squares of the errors** between predicted and actual values.
* It tells you **how wrong your modelâ€™s predictions are**, with **larger errors penalized more**.

---

### **2. Formula **
$$
MSE = \frac{1}{m} \sum_{i=1}^{m} \left( y^{(i)} - \hat{y}^{(i)} \right)^2
$$

---

**Where:**

- \(y^{(i)}\) = actual value  
- \(\hat{y}^{(i)}\) = predicted value  
- \(m\) = number of samples  


---

### **4. Key Points / Intuition**

* Squaring errors ensures **all differences are positive**.
* **Big errors are penalized more** than small errors (unlike MAE).
* Commonly used in **regression models** like Linear Regression.

---

### **5. Advantages**

1. Penalizes **large errors** more strongly â†’ good for sensitive applications.
2. Smooth and differentiable â†’ useful for **optimization (Gradient Descent)**.
3. Widely used and standard metric for regression.

---

### **6. Disadvantages**

1. Sensitive to **outliers** (a single large error can dominate MSE).
2. Harder to interpret in real units (squared units of target).

---

### **7. Summary (Simple)**

* **MSE = average squared difference between predicted and actual values**
* **Use it** when you want to penalize large errors more.
* **Donâ€™t use it** if your dataset has extreme outliers.

# **RMSE â€” Root Mean Squared Error**

### **1. Definition**

* RMSE is the **square root of the Mean Squared Error (MSE)**.
* It measures the **average magnitude of the error** between predicted and actual values.
* Unlike MSE, RMSE is in the **same units as the target variable**, making it easier to interpret.

> RMSE is commonly used in regression tasks and also as a **loss function** in Deep Learning.

---

### **2. Formula**


$$
RMSE = \sqrt{ \frac{1}{m} \sum_{i=1}^{m} \left( y^{(i)} - \hat{y}^{(i)} \right)^2 }
$$

---

### **3. Where**

**Where:**

* (y^{(i)}) = actual value
* (\hat{y}^{(i)}) = predicted value
* (m) = number of samples

---

### **4. Key Points / Intuition**

* Takes the **square root of MSE** â†’ makes the error interpretable in **original units**.
* Penalizes **large errors more** than small errors.
* Often preferred over MAE when **large errors are especially bad**.
* Smooth and differentiable â†’ can be used as a **loss function** in gradient-based optimization.

---

### **5. RMSE as a Loss Function**

* In **Deep Learning**, RMSE (or MSE) is often used as the **loss function** for regression tasks.
* **Why a loss function?**

  * The network predicts outputs.
  * RMSE measures **how far predictions are from true values**.
  * Gradient descent then **adjusts weights** to minimize RMSE.
* Essentially:

> RMSE is not just a metric; in DL it becomes the **signal that tells the network how to learn**.

---

### **6. Advantages**

1. Easy to interpret (same units as target).
2. Penalizes **large errors heavily**, which can be important in critical applications.
3. Smooth and differentiable â†’ works well with gradient descent.

---

### **7. Disadvantages**

1. Sensitive to **outliers** (big errors dominate).
2. Not as robust as MAE if dataset has noisy extreme values.

---

### **8. Summary**

* RMSE = sqrt(MSE)
* Used as **regression metric** and **loss function** in Deep Learning
* Tells **average prediction error in original units**
* Penalizes large errors more â†’ good for precise predictions

# **RÂ² Score â€” Explanation (Hands-On ML Style)**

### **1. Definition**

* RÂ² measures **how well your regression model predicts the target variable**.
* Specifically, it tells you the **proportion of the variance** in the target variable that is explained by the model.
* Value ranges from **âˆ’âˆž to 1**:

  * **1** â†’ perfect prediction
  * **0** â†’ model predicts as well as just the mean of y
  * **<0** â†’ model performs worse than predicting the mean

---

### **2. Formula**

$$
R^2 = 1 - \frac{SS_{\text{res}}}{SS_{\text{tot}}}
$$

---

### **Where:**

* (SS_{\text{res}} = \sum_i (y^{(i)} - \hat y^{(i)})^2) â†’ residual sum of squares (errors)
* (SS_{\text{tot}} = \sum_i (y^{(i)} - \bar y)^2) â†’ total variance of target
* (\bar y) = mean of actual target values

---

**Interpretation:**

> RÂ² = 1 â€“ (unexplained variance Ã· total variance)

---

### **3. Intuition**

* If your model **perfectly predicts** all targets â†’ residual errors = 0 â†’ RÂ² = 1
* If your model **predicts the mean for all points** â†’ residual errors = total variance â†’ RÂ² = 0
* If your model **predicts worse than the mean** â†’ residual errors > total variance â†’ RÂ² < 0

---

### **4. Key Points**

1. RÂ² gives a **normalized score of model fit** (percentage of variance explained).
2. Higher RÂ² â†’ better model fit.
3. Works for **linear and nonlinear regression**, but only for regression.
4. Donâ€™t use RÂ² alone; combine with **RMSE / MAE** for absolute error measurement.

---

### **5. Advantages**

* Provides **relative goodness-of-fit** metric
* Easy to **interpret** (like % of variance explained)
* Standard in regression tasks

---

### **6. Disadvantages / Cautions**

* Doesnâ€™t indicate **bias / overfitting / model correctness**
* Sensitive to **outliers** in some datasets
* Can be misleading for **nonlinear models** if used alone

---

### **7. Hands-On ML Tip**

* In Python / Scikit-Learn:

```python

print(r2)
```

* This gives you the **RÂ² score of your model on test data**.
* Always check alongside **RMSE or MAE** to understand absolute errors.

# **Adjusted RÂ² Score**

### **1. Definition**

* Adjusted RÂ² modifies the RÂ² score to account for the **number of features (predictors)** in your model.
* It **penalizes irrelevant features** that donâ€™t improve the model.
* Use it when you have **multiple regression** (more than one predictor).

> Intuition:

* Adding irrelevant features â†’ RÂ² may increase slightly (even if model doesnâ€™t actually improve)
* Adjusted RÂ² **decreases** if new features are irrelevant
* Adjusted RÂ² **increases** if new features are relevant

---

### **2. Formula **

$$
\text{Adjusted } R^2 = 1 - \left(1 - R^2\right) \frac{n - 1}{n - p - 1}
$$

Where:

* (R^2) = regular RÂ² score
* (n) = number of samples
* (p) = number of predictors (features)

---

### **3. Where **

* (R^2) = coefficient of determination (regular RÂ²)
* (n) = number of observations / samples
* (p) = number of features in the model

---

### **4. Key Points / Intuition**

1. Adjusted RÂ² **accounts for the number of predictors**, unlike regular RÂ².
2. Helps detect **overfitting**: adding irrelevant variables wonâ€™t falsely increase adjusted RÂ².
3. Always use in **multiple regression**, especially when comparing models with different numbers of features.
4. Can be **negative** if the model is very bad (even worse than predicting the mean).

---

### **5. Advantages**

* Penalizes irrelevant features â†’ avoids misleading RÂ² increases
* Better metric for **model comparison** in multiple regression
* Gives a more **realistic measure of fit** when you have many features

---

### **6. Disadvantages**

* Slightly more complex to interpret than regular RÂ²
* Only useful when you have **multiple predictors**
* Doesnâ€™t solve all problems: still sensitive to outliers

---

### âœ… **Summary**

* Adjusted RÂ² = RÂ² corrected for number of predictors
* Increases only if **new features improve the model**
* Decreases if **new features are irrelevant**
* Best used in **multiple regression**
